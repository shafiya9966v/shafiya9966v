```python
columns_to_drop=['id','cardio']
x=pd1.drop(columns='cardio',axis=1)
y=pd1['cardio']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy on training data
x_train_predict=model.predict(x_train)
training_dta_accuracy=accuracy_score(x_train_predict,y_train)
```

```python
print("accuracy on training data:",training_dta_accuracy)
```

```python
x_test_predict=model.predict(x_test)
test_dta_accuracy=accuracy_score(x_test_predict,y_test)
```

```python
print("accuracy on test data:",test_dta_accuracy)
```
